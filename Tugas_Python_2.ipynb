**Perhitungan Rata-rata, Standar deviasi, Skewness, Median, Modus, Energi, Min, Max**

```python
# Fungsi untuk menghitung statistik yang diminta dan menyimpannya di parameter dataset baru
def hitung_statistik(data):
    statistics = pd.DataFrame(columns=['Rata-rata', 'Standar Deviasi', 'Skewness', 'Median', 'Modus', 'Energi', 'Min', 'Max'])
    for col_idx, col_name in enumerate(data.columns):
        column_data = data[col_name]
        statistics.loc[col_idx] = [
            column_data.mean(),
            column_data.std(),
            skew(column_data),
            column_data.median(),
            column_data.mode().iloc[0],
            (column_data ** 2).mean(),
            column_data.min(),
            column_data.max()
        ]
    return statistics
```

```python
stats_file = hitung_statistik(isi_file)
stats_final = stats_file.T.rename(columns={col: f"Kolom {col+1}" for col in stats_file.T})
stats_final
```
